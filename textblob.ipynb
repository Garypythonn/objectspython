#Streaming en tiempo real

```python
# Bot que genera un streaming en tiempo real a partir de palabras clave:
from datetime import datetime

x=[]
class MyStream(tweepy.StreamingClient):
  
  # Esta funci√≥n s√≥lo se ejecuta si el streaming funciona:
  def on_connect(self):
      print("Connected")

    

  # This function gets called when a tweet passes the stream
  def on_tweet(self, tweet):
    # Displaying tweet in console
    if tweet.referenced_tweets == None:
          
      if "python" not in tweet.text:
        if "programming" not in tweet.text:
          if "coding" not in tweet.text:
            if "Python" not in tweet.text:
              if "#100DaysOfCode" not in tweet.text:
                if "Programming" not in tweet.text:
                  print(tweet.text)
                  x.append({'fecha':datetime.now(),'texto':tweet.text})
                  print("___________________________________________________________________________________________")
                  # Retraso entre tweets.
                  time.sleep(0.5)
                    
                

# Creamos una instancia de la clase MyStream:
stream = MyStream(bearer_token=bearer_token)

stream.add_rules(tweepy.StreamRule("petro"),dry_run=True)

    
# Inicializamos el stream
stream.filter(tweet_fields=["referenced_tweets"])
```

# Expresiones regulares: patrones para encontrar una determinada colecci√≥n de caracteres dentro de una cadena de texto
![](https://snipboard.io/nwFTr9.jpg) 

```python
#[] Se utiliza para indicar un conjunto de caracteres. En un conjunto

def limpiar_texto(text):
  text = re.sub(r'^RT[\s]+', '', text) #Quitamos los RT y los reemplazamos por espacios ''.
  text = re.sub(r'https?:\/\/.*[\r\n]*', '', text) #Quitamos los links.
  text = re.sub(r'#', '', text) #Quitamos los hashtags.
  text = re.sub(r'@[A-Za-z0-9]+', '', text) #Quitamos las menciones.
  return text
```

```python
data['clean_text']=data["texto"].apply(limpiar_texto) #Creamos una columna
data
```

```python
def obtener_polaridad(text):
  analysis = TextBlob(text)
  if text != '': #si el strin no es vac√≠o
    result = analysis.translate(from_lang = 'es', to = 'en').sentiment.polarity
    time.sleep(5)
    return result
```

```python
def obtener_subjetividad(texto):
  analisis = TextBlob(texto)
  if texto != '':
    resultado = analisis.translate(from_lang = 'es', to = 'en').sentiment.subjectivity
    time.sleep(5)
    return resultado
```

```python
print(obtener_polaridad("te odio"))
#print(obtener_subjetividad("te odio"))
```

```python
data['polaridad'] = data['clean_text'].apply(obtener_polaridad)
data['subjetividad'] = data['clean_text'].apply(obtener_subjetividad)

```

```python
#Creamos una funci√≥n para dejar la polaridad en forma "binaria"
def x_range(x):
  if x > 0:
    return 1
  elif x == 0:
    return 0
  else:
    return -1
```

```python
data['polaridadBinario'] = data['polaridad'].apply(x_range)
data.head(3)
```

```python
#Resumen de la columna de binarios
data['polaridadBinario'].value_counts()
```
